```python
import numpy as np
from numpy.linalg import norm
from math import sqrt

def m1(u, g, a ):

    a1, a2 = a
    g1, g2 = g

    m = a1*(sqrt((1+u[0])**2+(1+u[1])**2)-sqrt(2))**2 + a2*(sqrt((1-u[0])**2+(1+u[1])**2)-sqrt(2))**2 - g1*u[0] - g2*u[1]
    return m

def mCN(u,kn,c):
    gn = u[1]-c
    return 0.5*kn*gn**2 if gn>0 else 0

def mCT(u,kt,p):
    gt = u[0]-p
    return 0.5*kt*gt**2

def dm1(u, g, a ):
    return [2*a[0]*(u[0] + 1)*(sqrt((u[0] + 1)**2 + (u[1] + 1)**2) - sqrt(2))/sqrt((u[0] + 1)**2 + (u[1] + 1)**2) + 2*a[1]*(u[0] - 1)*(sqrt((1 - u[0])**2 + (u[1] + 1)**2) - sqrt(2))/sqrt((1 - u[0])**2 + (u[1] + 1)**2) - g[0],
            2*a[0]*(u[1] + 1)*(sqrt((u[0] + 1)**2 + (u[1] + 1)**2) - sqrt(2))/sqrt((u[0] + 1)**2 + (u[1] + 1)**2) + 2*a[1]*(u[1] + 1)*(sqrt((1 - u[0])**2 + (u[1] + 1)**2) - sqrt(2))/sqrt((1 - u[0])**2 + (u[1] + 1)**2) - g[1]]

def fCN(u,kn,c):
    gn = u[1]-c
    dmdgn = -kn*gn if gn>0 else 0
    dgndu = np.array([0,1])
    return dmdgn*dgndu

def fCT(u,kt,p):
    gt = u[0]-p
    dmdgt = -kt*gt
    dgtdu = np.array([1,0])
    return dmdgt*dgtdu

def d2m1(u,g,a):
    return [[2*(a[0]*(u[0] + 1)**2/((u[0] + 1)**2 + (u[1] + 1)**2) - a[0]*(u[0] + 1)**2*(sqrt((u[0] + 1)**2 + (u[1] + 1)**2) - sqrt(2))/((u[0] + 1)**2 + (u[1] + 1)**2)**(3/2) + a[0]*(sqrt((u[0] + 1)**2 + (u[1] + 1)**2) - sqrt(2))/sqrt((u[0] + 1)**2 + (u[1] + 1)**2) + a[1]*(1 - u[0])*(u[0] - 1)*(sqrt((1 - u[0])**2 + (u[1] + 1)**2) - sqrt(2))/((1 - u[0])**2 + (u[1] + 1)**2)**(3/2) + a[1]*(u[0] - 1)**2/((1 - u[0])**2 + (u[1] + 1)**2) + a[1]*(sqrt((1 - u[0])**2 + (u[1] + 1)**2) - sqrt(2))/sqrt((1 - u[0])**2 + (u[1] + 1)**2)), 2*(u[1] + 1)*(a[0]*(u[0] + 1)/((u[0] + 1)**2 + (u[1] + 1)**2) - a[0]*(u[0] + 1)*(sqrt((u[0] + 1)**2 + (u[1] + 1)**2) - sqrt(2))/((u[0] + 1)**2 + (u[1] + 1)**2)**(3/2) + a[1]*(1 - u[0])*(sqrt((1 - u[0])**2 + (u[1] + 1)**2) - sqrt(2))/((1 - u[0])**2 + (u[1] + 1)**2)**(3/2) + a[1]*(u[0] - 1)/((1 - u[0])**2 + (u[1] + 1)**2))], [2*(u[1] + 1)*(a[0]*(u[0] + 1)/((u[0] + 1)**2 + (u[1] + 1)**2) - a[0]*(u[0] + 1)*(sqrt((u[0] + 1)**2 + (u[1] + 1)**2) - sqrt(2))/((u[0] + 1)**2 + (u[1] + 1)**2)**(3/2) + a[1]*(u[0] - 1)/((1 - u[0])**2 + (u[1] + 1)**2) - a[1]*(u[0] - 1)*(sqrt((1 - u[0])**2 + (u[1] + 1)**2) - sqrt(2))/((1 - u[0])**2 + (u[1] + 1)**2)**(3/2)), 2*(a[0]*(u[1] + 1)**2/((u[0] + 1)**2 + (u[1] + 1)**2) - a[0]*(u[1] + 1)**2*(sqrt((u[0] + 1)**2 + (u[1] + 1)**2) - sqrt(2))/((u[0] + 1)**2 + (u[1] + 1)**2)**(3/2) + a[0]*(sqrt((u[0] + 1)**2 + (u[1] + 1)**2) - sqrt(2))/sqrt((u[0] + 1)**2 + (u[1] + 1)**2) + a[1]*(u[1] + 1)**2/((1 - u[0])**2 + (u[1] + 1)**2) - a[1]*(u[1] + 1)**2*(sqrt((1 - u[0])**2 + (u[1] + 1)**2) - sqrt(2))/((1 - u[0])**2 + (u[1] + 1)**2)**(3/2) + a[1]*(sqrt((1 - u[0])**2 + (u[1] + 1)**2) - sqrt(2))/sqrt((1 - u[0])**2 + (u[1] + 1)**2))]]


def mN(u,g,a,kn,c):
    return m1(u,g,a) + mCN(u,kn,c)
def mT(u,g,a,kt,p):
    return m1(u,g,a) + mCT(u,kt,p)
def fN(u,g,a,kn,c):
    return dm1(u,g,a) + fCN(u,kn,c)
def fT(u,g,a,kt,p):
    return dm1(u,g,a) + fCT(u,kt,p)
```

```python
def STEEPDESC(funct, dfunct, a, g, u0, kn,kt,yc,xp, alpha = 0.04, BTLS = False, CG = True, c=0.5, r=0.5, n_reset = 20, BetaIdx = 1 ):
    import time

    if funct == mN:
        param1 = kn
        param2 = yc
    elif funct == mT:
        param1 = kt
        param2 = xp

    
    ti = time.time()
    m  =  funct
    dm = dfunct
    u = u0
    m_new = m(u,g,a,param1,param2)
    m_old = 10**100
    f_new = np.zeros_like(u)
    h_new = np.zeros_like(u)
    iter = 0
    subiter = 0

    while m_new < m_old:
        m_old = m_new
        f_old = f_new
        h_old = h_new

        f_new = np.array(dm(u,g,a,param1,param2))
        # h = -f

        if iter%n_reset==0:
            h_new = -1*f_new
        else:
            if BetaIdx == 1:
                beta = np.dot(f_new,f_new)/np.dot(f_old,f_old)
            elif BetaIdx ==2:
                beta = np.dot(f_new,f_new-f_old)/np.dot(f_old,f_old)
            elif BetaIdx ==3:
                beta = np.dot(f_new,f_new-f_old)/np.dot(h_old,f_new-f_old)
            elif BetaIdx ==4:
                beta = np.dot(f_new,f_new)/np.dot(h_old,f_new-f_old)

            h_new = -1*f_new + max([0,beta])*h_old

        if BTLS:
            alpha = 1/r*alpha
            mx = 10**100
            while mx>m_new+c*alpha*np.dot(h_new,f_new):
                alpha = r*alpha     #to make it go back to its initial value (in the first iteration)
                ux = u + alpha*h_new
                mx = m(ux,g,a)
                subiter += 1

            # else:
            #     if iter>16:
            #         set_trace()

            u = ux    
            m_new = mx

        else:
            u += alpha*h_new
            m_new = m(u,g,a,param1,param2)

        iter += 1

    # print("Steepest descent takes "+str(iter)+" iterations and "+str(subiter)+" subiterations for alpha = "+str(alpha))
    return u, m_old, iter, time.time()-ti
```
