```python
import numpy as np
from matplotlib import pyplot as plt
from math import sqrt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

a, ha, b, hb, hc, hd  = [4, -1, 4/sqrt(2), 2, 3.5, -2.5]
X = np.array([[ a, 0,ha], [ 0, a,ha], [-a, 0,ha], [ 0,-a,ha], [ b, b,hb],
              [-b, b,hb], [-b,-b,hb], [ b,-b,hb], [ 0, 0,hc], [ 0, 0,hd]])
C = [[1,2,5], [ 2, 3, 6], [ 3, 4, 7], [ 4, 1, 8], [ 1, 5, 8], [ 2, 6, 5], [ 3, 7, 6], [ 4, 8, 7], 
     [5,9,8], [ 6, 9, 5], [ 7, 9, 6], [ 8, 9, 7], [ 1,10, 2], [ 2,10, 3], [ 3,10, 4], [ 4,10, 1]]

vol = 0
for (i,j,k) in C:
    ap,p,bp = X[i-1], X[j-1], X[k-1]
    c = np.cross(ap-p,bp-p)
    vol += np.dot(p,c)/6

triangles = [((X[i-1],X[j-1],X[k-1])) for (i,j,k) in C]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.add_collection(Poly3DCollection(triangles,facecolor=(0,0,1,0.5),edgecolor="black"))
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.set_zlim([-4,4])
plt.show()
vol
```
